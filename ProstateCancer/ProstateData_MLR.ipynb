```python
# IMPORT DATA :

import pandas as pd
from sklearn import linear_model
from sklearn.model_selection import train_test_split


df =pd.read_csv("drive/My Drive/IntroML/ProstateCancer/ProstateData_post_Treatment.csv")
df.head()
```

For this database we have chosen to focus on explaining MEDV.

```python
# Prepare Data :

x = df.drop('lpsa', axis=1)
y = df['lpsa']
 
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.3, random_state=42)
```

```python
# Multi Linear Model :
# with sklearn
regr = linear_model.LinearRegression()
regr.fit(X_train, y_train)

print('Intercept: \n', regr.intercept_)
print('Coefficients: \n', regr.coef_)
```

```python
# Graph :
from matplotlib import pyplot as plt 
import seaborn as sns 


y_pred = regr.predict(X_test)
plt.scatter(y_test, abs(y_pred - y_test) ,color='g') 



plt.show()
```

```python
# RMSE :

from sklearn.metrics import mean_squared_error

mean_squared_error(y_test, y_pred)
```
