```python
Amplitude = max(y) - min(y)
print("L'amplitude est de : ",Amplitude)
Errors_normalized = Errors / Amplitude
plt.figure(figsize=(8, 6), dpi=80)
plt.subplot(2,1,1)
plt.bar(Models,Errors_normalized,label = "SME")
plt.ylabel("SME normaliséé par l'amplitude")
plt.legend()
plt.subplot(2,1,2)
plt.bar(Models,Times,label= "Computational time")
plt.ylabel("t en s")
plt.yscale('log')
plt.legend()
```
