```python
import pandas as pd
import numpy as np
import time
from matplotlib import pyplot as plt 
```

```python
Models = ['Forest_model','MLR_model','MLP_model','PCA_linear_model','SVM_model']

Errors_normalized_Prostate_Cancer = np.array([0.07458078, 0.07670732, 0.31450801, 0.12861901, 0.76084078])
Times_Prostate_Cancer= np.array([21.76865863800049,0.011000871658325195,0.1490309238433838,0.004999876022338867, 0.009000539779663086])
Shape_Prostate_Cancer = [8,97]
Times_normalized_Prostate_Cancer = Times_Prostate_Cancer/(Shape_Prostate_Cancer[0]*Shape_Prostate_Cancer[1])

Errors_normalized_Housing = np.array([0.12031514, 0.48485462, 0.23817541, 1.3000708 , 1.37743275])
Times_Housing = np.array([31.529433488845825,0.01600360870361328,0.7361650466918945,0.0039997100830078125,0.07817220687866211])
Shape_Housing =  [13,506]
Times_normalized_Housing = Times_Housing/(Shape_Housing[0]*Shape_Housing[1])

Errors_normalized_Dataset_Dasci = np.array([0.00215186, 0.00266523, 0.00267729, 0.03611389, 0.02713622])
Times_Dataset_Dasci= np.array([324.88870215415955,0.013534307479858398,5.115051031112671,0.016071319580078125,13.949984073638916])
Shape_Dataset_Dasci =  [17,8566]
Times_normalized_Dataset_Dasci = Times_Dataset_Dasci/(Shape_Dataset_Dasci[0]*Shape_Dataset_Dasci[1])


```

```python
# Position sur l'axe des x pour chaque étiquette
position = np.arange(len(Models))
# Largeur des barres
largeur = .20

plt.figure(figsize=(8, 6), dpi=80)
#plt.bar(Models,Errors_normalized_Prostate_Cancer,Errors_normalized_Housing,label = ["Prostate_cancer","Housing"])
fig, ax = plt.subplots(figsize=(8, 6))
r1 = ax.bar(position - 3*largeur/2, Errors_normalized_Prostate_Cancer, largeur,label="Prostate Cancer")
r2 = ax.bar(position - largeur/2, Errors_normalized_Housing, largeur,label="Housing")
r3 = ax.bar(position + largeur/2, Errors_normalized_Dataset_Dasci, largeur,label="Dasci Data_set")

# Modification des marques sur l'axe des x et de leurs étiquettes
ax.set_xticks(position)
ax.set_xticklabels(Models)
plt.ylabel("SME normaliséé par l'amplitude")
ax.legend()




#plt.subplot(2,1,2)
#plt.bar(Models,Times_Prostate_Cancer,Times_Housing,label = ["Prostate_cancer","Housing"])
#plt.ylabel("computational time en s")
#plt.legend()
```
